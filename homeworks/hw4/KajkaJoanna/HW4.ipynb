```python
names = ["nikt konkretny", "nie wiem", "podmioty zagraniczne, np Rosja", "przedsiębiorstwa produkujące energię", "Unia Europejska", "rząd"]
values = [1 ,8, 19, 28, 28, 63]
```

1. Dlaczego wykres jest lepszy od wykresu kołowego:
Wykresy kołowe z reguły są mniej czytelne, ze względu na często mylące kąty. Dodatkowo wykres ten ma nielogicznie ułożone wartości, któych suma nie wynosi 100, co w wykresie kołowym jest niedopuszczlne. Dlatego lepszym rozwiązaniem jest zwizualizowanie danych za pomocą wykresu słupkowego.
2. Dlaczego wykres jest lepszy od wykresu z pracy domowej:
Na wykresie z pracy domowej każdy słupek jest w innym kolorze, co nie jest poprawne. Dlatego w wykresie w Pythonie zastosowałam jeden wyraźny kolor. Dodatkowo w poprawionym wykresie wartości są ułożone malejąco, a nie niemonotonicznie jak w wykresie w ggplot2.
