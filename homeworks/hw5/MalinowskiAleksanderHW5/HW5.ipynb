```python
import plotly.graph_objects as go
import csv
import numpy as np
import pandas as pd
from jupyter_dash import JupyterDash
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
df = pd.read_csv("complete.csv")
df_transformed = df[["awardYear","category","prizeAmount","prizeAmountAdjusted","fullName","gender","birth_country"]]
df_transformed = df_transformed.loc[(df_transformed.prizeAmountAdjusted > 5000000) & (df_transformed.birth_country != "")]
df2 = df_transformed.groupby("birth_country")['category'].value_counts().reset_index(level=[0]).rename(columns = {"category":"count"}).reset_index()
categories = df2.iloc[::,1].unique().tolist()
```

```python
from dash.dependencies import Input, Output
import random
import plotly.express as px
indices = [random.randrange(len(categories)) for i in range(5)]
app = JupyterDash(__name__)
app.layout = html.Div([
    dcc.Dropdown(
        id = "dropdown",
        options = [{"label":x,"value":x} for x in categories],
        value = [categories[i] for i in indices],
        multi = True
    ),
    dcc.Graph(id="chart"),
])

@app.callback(
    Output("chart","figure"),
    [Input("dropdown","value")])
def update_bar_chart(x):
    df_temp = df2.loc[df2["birth_country"].isin(x)]
    fig = px.bar(
        data_frame = df_temp,y = df_temp['count'],x = df_temp['birth_country'] , color = df_temp["category"], barmode = "group", title = "Liczba zdobywców nagrody Nobla w zależności od państwa urodzenia w podziale na kategorie")
    return fig
app.run_server(debug=True,mode="jupyterlab")
```
