```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("~/Downloads/complete.csv", keep_default_na = False)
df["continent"] = df.birth_continent + df.org_founded_continent
```

```python
fig = px.strip(df, x = "awardYear", y = "category",
 color = "continent",
 stripmode= "group", 
 hover_name = "name",
 title = "Where the Nobel Prize winners come from?",
 labels = {"category": "Category", "awardYear": "Year of receiving the award", "continent" : "Continent of origin"})
fig
```

```python
fig.write_html("./wykres.html")
```
